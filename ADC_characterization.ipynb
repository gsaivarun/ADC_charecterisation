```python
x = [6,7,8,9,10]
y_ch1 = [1.76, 2.20, 2.78, 3.53, 3.75]
plt.plot(x, y_ch1, label=f'ch1', color='b', marker = 's')

y_ch2 = [1.74, 2.21, 2.77, 3.51, 3.76]
plt.plot(x, y_ch2, label=f'ch2', color='g', marker = 'o')

y_ch3 = [1.75, 2.22, 2.78, 3.54, 3.78]
plt.plot(x, y_ch3, label=f'ch3', color='r', marker = 'v')

y_ch4 = [1.74, 2.23, 2.75, 3.52, 3.79]
plt.plot(x, y_ch4, label=f'ch4', color='cyan', marker = '^')

y_ch5 = [1.77, 2.23, 2.77, 3.50, 3.78]
plt.plot(x, y_ch5, label=f'ch5', color='yellow', marker = '>')

y_ch6 = [1.76, 2.21, 2.78, 3.53, 3.77]
plt.plot(x, y_ch5, label=f'ch6', color='black', marker = '<')

y_ch7 = [1.74, 2.22, 2.78, 3.53, 3.78]
plt.plot(x, y_ch7, label=f'ch7', color='pink', marker = '.')

y_ch8 = [1.74, 2.22, 2.78, 3.53, 3.78]
plt.plot(x, y_ch8, label=f'ch8', color='grey', marker = '*')

plt.xlabel('Power(dBm)')
plt.ylabel('Voltage')
plt.yscale('linear')
plt.title('ADC1-Voltage vs Power')
plt.grid(True)
plt.legend()
plt.show()


list_of_lists = [y_ch1, y_ch2, y_ch3, y_ch4, y_ch5, y_ch6, y_ch7, y_ch8]

first_elements = []
second_elements = []
third_elements = []
fourth_elements = []
fifth_elements = []

for sublist in list_of_lists:
    first_elements.append(sublist[0])
    second_elements.append(sublist[1])
    third_elements.append(sublist[2])
    fourth_elements.append(sublist[3])
    fifth_elements.append(sublist[4])


avg1= np.average(first_elements)
avg2= np.average(second_elements)
avg3= np.average(third_elements)
avg4= np.average(fourth_elements)
avg5= np.average(fifth_elements)

y_avg = [avg1,avg2,avg3,avg4,avg5]


y1 = np.std(first_elements)
y2 = np.std(second_elements)
y3 = np.std(third_elements)
y4 = np.std(fourth_elements)
y5 = np.std(fifth_elements)

y_error = [y1, y2, y3, y4, y5]


plt.errorbar(x, y_avg, yerr=y_error, label='ch', color='r', marker='.', capsize=5)

plt.xlabel('Power(dBm)')
plt.ylabel('Avg Voltage of channels at a perticular power')
plt.yscale('linear')

plt.title('Error bars ADC1-Voltage vs Power')
plt.grid(True)

plt.legend()

plt.show()
```
