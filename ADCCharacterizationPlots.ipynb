```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = 'ADC1.csv'
df = pd.read_csv(file_path)

x_values = df.iloc[:, 0].tolist()
channels = df.columns[1:]
y_values = [df[channel].tolist() for channel in channels]

for channel in channels:
    plt.plot(x_values, df[channel], marker='o', label=channel)
plt.xlabel('Power(dBm)')
plt.ylabel('Voltage')
plt.title('ADC1 - Voltage vs Power')
plt.grid(True)
plt.legend()
plt.show()

row_averages = df.iloc[:, 1:].mean(axis=1)
plt.plot(x_values, row_averages, linestyle='-', color='black')

y_error = df.iloc[:, 1:].std(axis = 1)

plt.errorbar(x_values, row_averages, yerr=y_error, label='Average', color='r', marker='.', capsize=5)

plt.xlabel('Power(dBm)')
plt.ylabel('Avg Voltage of channels at a particular power')
plt.yscale('linear')
plt.title('Error bars ADC1-Voltage vs Power')
plt.grid(True)
plt.legend()
plt.show()
print(y_error)
```
